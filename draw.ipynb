## Reading csv file

```python
df =pd.read_csv('currencydata.csv')
```

```python
def filtercurrency(msg):
    msg=msg.split("=")
    dollor=msg[0]
    dollor=dollor.replace("US Dollar","").strip()
    dollor=float(dollor.replace("USD","").strip())
    rupee=msg[1]
    rupee=float(rupee.replace("INR","").strip())
    return (dollor,rupee)
```

```python
def filtertime(msg):
    tim=datetime.datetime.strptime(msg,"%Y-%m-%d %H:%M:%S")
    return tim.strftime("%Y-%m-%d %H:%M:%S")
    # X.append(tim.strftime("%H:%M"))
```

```python
def process(msg):
    return msg[1]
```

```python
df["Date"]=df["insertime"].apply(filtertime)
df['currencyinfo']=df['currencyinfo'].apply(filtercurrency)
df['dollor']=df['currencyinfo'].apply(process)
```

```python
dollr=df['dollor']
```

## ploting graph of USD with respect ot time 

```python
plt.figure(figsize=(10, 6), dpi=150, facecolor='g', edgecolor='k')
dollr.plot(title="USD-INR time Graph")
plt.ylabel('1 USD value in INR');
plt.xlabel('USD value in 10 min interval');
fig = plt.gcf()
fig.savefig('graph1.png')
```
